Recuerda que:


*   Una **clave primaria** es un atributo (o conjunto) que identifica unívocamente a cada registro en la tabla.
*   Una **clave foránea** (externa o ajena) es un atributos (o conjunto) en una tabla que es una clave primaria en otra (o posiblemente la misma) tabla.
*   Las **relaciones** son las líneas que conectan una tabla con otra y el extremo determina la cardinalidad. Las relaciones con línea continua (identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser foránea y primaria (al mismo tiempo) de otra. Las relaciones con línea discontinua (no identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser sólo foránea en otra.

11.	Carga las tablas empleadas en dataframes con el mismo nombre y resuelve las consultas anteriores con las funciones de Pandas. Cuida no sobreescribir los dataframes originales al resolver las consultas. Debes obtener los mismos resultados que con SQL.

```python
productlines = pd.read_sql("Select * from productlines", db)
products = pd.read_sql("Select * from products", db)
orderdetails = pd.read_sql("Select * from orderdetails", db)
orders = pd.read_sql("Select * from orders", db)
employees = pd.read_sql("Select * from employees", db)
customers = pd.read_sql("Select * from customers", db)
offices = pd.read_sql("Select * from offices", db)
payments = pd.read_sql("Select * from payments", db)
display(productlines)
display(employees.sort_values(by=['firstName']))
display(offices.filter(items=['country']).drop_duplicates())
display(customers.filter(items=['customerName','phone']).loc[customers['city']=='NYC'])
display(products.filter(items=['productCode','productName']).loc[(products['productVendor']=='Gearbox Collectibles') & (products['quantityInStock']<1000)])
display(products.sort_values(by=['buyPrice'], ascending=False).head(3))
display(products.groupby(['productLine']).size())
display(employees.set_index('officeCode').join(offices.set_index('officeCode')).groupby(['country']).size())
display(payments.set_index('customerNumber').join(customers.loc[customers['country']=='Spain'].set_index('customerNumber')).groupby(['country']).mean('amount'))
```
